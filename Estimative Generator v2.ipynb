```python
import pandas as pd

headers = responseheaders['values'][0]
values = response['values'][0]



wallets = {}



def CreateWallet(name,amount=0,desc='',annualrate=0, wallets={}):
    wallets[name]= {
    'amount':amount,
    'name':name,
    'desc':desc,
    'annualrate':annualrate,
    'startdate':'',
    'enddate':''}
        
    return wallets

    
#populate with sheets
for h,v in zip(headers,values):
    #print(h,v)
    wallets = CreateWallet(name=h,amount=v)
    
#set annual rates
wallets['Banco CTT']['annualrate']=0
wallets['Moey']['annualrate']=0
wallets['Bni Europa Poup']['annualrate']=0
wallets['Revolut EUR']['annualrate']=0
wallets['Paypal (EUR)']['annualrate']=0
wallets['ActivoBank']['annualrate']=0
wallets['OnHand']['annualrate']=0

#set anual rates investing wallets
wallets['Bni Europa Corr']['annualrate']=0.01
wallets['Revolut Invest in EUR(1 Dollar = 0.88 EUR<)']['annualrate']=0.06
wallets['Mintos']['annualrate']=0.05
wallets['Trading 212']['annualrate']=0.04
wallets['Degiro Basic']['annualrate']=0.03
wallets['XTB']['annualrate']=0.02
wallets['Revolut BitCoin']['annualrate']=0.02
wallets['Bankinter']['annualrate']=0.05
wallets['Bondora']['annualrate']=0.0675
wallets['Degiro Custody']['annualrate']=0.09


#quando gastas ao mes
monthlyspending=1500

amounts={}
#expected montly increase in each wallet - reforços mensais
amounts['Banco CTT']=0
amounts['Moey']=1310.7  -600-600-50 #salario - o sitio onde o vou investir
amounts['Bni Europa Corr']=0
amounts['Bni Europa Poup']=0
amounts['Revolut EUR']=600 #
amounts['Revolut Invest in EUR(1 Dollar = 0.88 EUR<)']=0
amounts['Mintos']=0
amounts['Paypal (EUR)']=30 #unity
amounts['Bondora']=0
amounts['Degiro Custody']=600
amounts['Revolut BitCoin']=50
amounts['Bankinter']=0
amounts['ActivoBank']=0
amounts['XTB']=0
amounts['Trading 212']=0
amounts['Degiro Basic']=0
amounts['OnHand']=0

```

```python
import datetime
import altair as alt



today = datetime.datetime.today()

#qunado vais morrer - usada p simular a quantia que precisas para viver ate la
dayidie = datetime.datetime(2086,1,1)


lifetime = dayidie-today
#months till i die
monthstolive = int(lifetime.days/30.2)

data={
    'moneyneededtolive':[],
    'networth':[],
    'x':[]
}

yearstocalculate = 30


seriesofmonths = range(1,monthstolive)

#set up initial investmen

walletmoneys={}

#for every month in the next  yearstocalculate
for i in seriesofmonths:
    
    if(i/12>yearstocalculate):
        break
    
    #ano - 24 corresponde  a minha idade
    data['x'].append(24+i/12)#i)
    
    #quantia pa viver
    data['moneyneededtolive'].append((monthlyspending)*(monthstolive-i))
  
    #assume capitalizacao anual (juro simples mensal)
    allmoneys=0.0
    for wallet_name in wallets:
        #dinheiro deste mes equivale + dinheiro que vou adicionar a cada mes, multiplacdo pelo juro
        
        walletmoneys[wallet_name] = (wallets[wallet_name]['amount']+i*amounts[wallet_name])*(1
               +i*(wallets[wallet_name]['annualrate']/12))
        #print(i,wallet_name,(wallets[wallet_name]['amount']+i*amounts[wallet_name])*(i*(annualrate/12)))
        allmoneys=allmoneys + walletmoneys[wallet_name]
    
    #remove dinheiro que vou gastar nesse mes
    #allmoneys= allmoneys - monthlyspending

    #adiciona quantia do mes à timeseries final
    data['networth'].append(allmoneys)
        
    
import pandas as pd
    
#d = {'x': seriesofyears, 'Needed Money': moneyneededtolive,'Net Worth':networth}
#df = pd.DataFrame(data=d)
df=pd.DataFrame.from_dict(data)

dfmelted= df.melt('x', var_name='Series', value_name='y')
```

```python


source=dfmelted

# Create a selection that chooses the nearest point & selects based on x-value
nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['x'], empty='none')



# The basic line
line = alt.Chart(source).mark_line().encode(
    x='x:Q',
    y='y:Q',
    color='Series:N'
)

# Transparent selectors across the chart. This is what tells us
# the x-value of the cursor
selectors = alt.Chart(source).mark_point().encode(
    x='x:Q',
    opacity=alt.value(0),
).add_selection(
    nearest
)

# Draw points on the line, and highlight based on selection
points = line.mark_point().encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

# Draw text labels near the points, and highlight based on selection
text = line.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(nearest, 'y:Q', alt.value(' '))
)

# Draw a rule at the location of the selection
rules = alt.Chart(source).mark_rule(color='gray').encode(
    x='x:Q',
).transform_filter(
    nearest
)

# Put the five layers into a chart and bind the data
alt.layer(
    line, selectors, points, rules, text
).properties(
    width=600, height=300
)
```
